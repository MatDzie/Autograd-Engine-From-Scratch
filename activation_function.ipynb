## Activation Function example

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# tanh
plt.plot(np.arange(-5, 5, 0.1), np.tanh(np.arange(-5, 5, 0.1)))
```

```python
# sigmoid
plt.plot(np.arange(-5, 5, 0.1), 1/(1+np.exp(-np.arange(-5, 5, 0.1))))
```

```python
# relu
plt.plot(np.arange(-5, 5, 0.1), np.maximum(np.arange(-5, 5, 0.1), 0))
```

```python
# softmax
plt.plot(np.arange(-5, 5, 0.1), np.exp(np.arange(-5, 5, 0.1))/np.sum(np.exp(np.arange(-5, 5, 0.1))))
```
